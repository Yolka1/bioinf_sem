```python
#Калькулятор
s = input('Введите ваше число: (используйте пробелы между символами) ')
parts = s.split()
a = int(parts[0])
operator = parts[1]
b = int(parts[2])
if operator == '+':
  print('Результат: ', a+b)
elif operator == '-':
  print('Результат: ', a-b)
elif operator == '*':
  print('Результат: ', a*b)
elif operator == '/':
  if b != 0:
    print('Результат: ', a/b)
  else:
    print ('На ноль делить нельзя')
```

```python
#Фигура
n = 7
for i in range(n):        # Внешний - строки 
    for j in range(n):    # Внутренний - столбцы
        # Если первая или последняя ИЛИ номер столбца = номер строки ИЛИ обратная диагональ
        if i == 0 or i == n - 1 or i == j or i + j == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()  
```

```python
#Вывод простых чисел
start = int(input("От: "))
end = int(input("До: "))

for num in range(start, end + 1):
    if num < 2: # 0, 1 (отрицательные) - непростые 
        continue
        
    prostoe = True 
    
    for i in range(2, num): 
        if num % i == 0:
            prostoe = False 
            break 
            
    if prostoe:
        print(num)
```

```python
#Вклад
x = int(input()) 
p = int(input()) 
y = int(input()) 
years = 0
while x < y:
    x = int(x + x * p / 100)
    years += 1
print(years)
```

```python
#Вывод второго самого большого
max1 = 0 # Самое большое
max2 = 0 
while True:
    n = int(input())
    if n == 0:
        break
    
    if n > max1:
        max2 = max1 
        max1 = n    
    elif n > max2:
        max2 = n     
print(max2)
```
