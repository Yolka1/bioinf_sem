```python
text = input("Введите текст: ")
alphabet = "abcdefghijklmnopqrstuvwxyz"
shifted_alphabet = alphabet[1:] + alphabet[:1]
table = str.maketrans(alphabet, shifted_alphabet) #карта замен
result = text.translate(table)
print(result)
```
