В настольной игре Скрабл (Scrabble) каждая буква имеет определенную ценность. Напишите программу, которая вычисляет стоимость введенного пользователем слова. Будем считать, что на вход подается только одно слово, которое содержит либо только английские, либо только русские буквы.

```python
points = {
    **dict.fromkeys("AEIOULNSTRАВЕИНОРСТ", 1), #вынимаем из составленного словаря и вставляем в наш
    **dict.fromkeys("DGДКЛМПУ", 2),
    **dict.fromkeys("BCMPБГЁЬЯ",3),
    **dict.fromkeys("FHVWYЙЫ",4),
    **dict.fromkeys("KЖЗХЦЧ",5),
    **dict.fromkeys("JXШЭЮ",8),
    **dict.fromkeys("QZФЩЪ",10),
}
word = input("Введите слово: ").upper()
total_score = 0 
for letter  in word:
    score = points.get(letter, 0) #ищет букву, в противном случае будет 0
    total_score += score
print(f"Стоимость слова: {total_score}")
```

В базе данных ветеринарной клиники информация о пациентах-котах хранится в списке кортежей. Данные о кошках и их владельцах записаны в формате «Кличка животного, Возраст животного, Имя владельца, Фамилия владельца»ю

Обнаружилось, что имена некоторых владельцев повторяются, потому что у них несколько кошек. Необходимо оптимизировать хранение данных таким образом, чтобы для каждого владельца при выводе на печать данные о всех его животных отображались в одной строке: Игорь Бероев: Муся, 7; Изольда, 2

```python
cat_data = [
    ('Муся', 7, 'Игорь', 'Бероев'),
    ("Изольда", 2, "Игорь", "Бероев"),
    ("Барсик", 5, "Анна", "Смирнова"),
    ("Борщ", 4, "Анна", "Смирнова"),  
    ("Лобзик", 1, "Дмитрий", "Иванов"),
    ("Мокасин", 3, "Елена", "Петрова"),
    ("Четверг", 2, "Дмитрий", "Иванов"), 
    ("Барбарис", 8, "Игорь", "Бероев"),    
    ("Интеграл", 6, "Елена", "Петрова"),
    ("Гипофиз", 7, "Алексей", "Сидоров")
]
storage = {}

for cat_name, age, owner_name, owner_surname in cat_data:
    
    full_name = f"{owner_name} {owner_surname}"
    cat_info = f"{cat_name} {age}"

    if full_name not in storage:
        storage[full_name] = []

    storage[full_name].append(cat_info)

for owner, cats in storage.items():
    all_cats = "; ".join(cats)
    print(f"{owner}: {all_cats}")
```

Сортировка пузырьком.
Сгенерируйте случайный массив длиной N=10. Отсортируйте его с помощью алгоритма сортировки пузырьком
Сгенерируйте массивы разной длины (10,100,1000,10000, 100000). Замерьте время работы сортировки пузырьком. Сделайте вывод о времени работы алгоритма.

```python
import random
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

lengths = [10, 100, 1000, 10000]

for N in lengths:
    test_list = [] 
    for i in range(N):
        random_number = random.randint(1, 1000) # random integer
        test_list.append(random_number) # кладем в список
    
    start_time = time.time() #Unix Epoch
    bubble_sort(test_list)
    end_time = time.time()
    duration = end_time - start_time
    
    print(f"Массив из {N} элементов отсортирован за {duration:.4f} секунд")

```

Реализуйте детскую игру “Ханойская башня”. Вместо колечек разной длины - цифры от 1 до 7. Игра инициализируется в случайном порядке. Игрок имеет три колышка, между которыми он может перемещать колечки. Игрок указывает с какого колышка на какой переместить верхнее колечко. Цель - собрать упорядоченную башенку на одном из колышков.

```python
import random

def play_hanoi():
    # 1. Инициализация игры
    pegs = {
        '1': [],
        '2': [],
        '3': []
    }

    
    rings = [7, 6, 5, 4, 3, 2, 1]
    
    # Раскидываем кольца по колышкам
    for ring in rings:
        random_peg = str(random.randint(1, 3))
        pegs[random_peg].append(ring)

    target_tower = [7, 6, 5, 4, 3, 2, 1] # Эталон

    print("--- ДОБРО ПОЖАЛОВАТЬ В ХАНОЙСКУЮ БАШНЮ ---")
    print("Правило: нельзя класть большее число на меньшее.")
    print("Цель: собрать башню [7, 6, 5, 4, 3, 2, 1] на 2-м или 3-м колышке.")

    # 2. Игровой цикл
    while True:
        # Рисуем текущее состояние
        print("\nТЕКУЩЕЕ СОСТОЯНИЕ:")
        for name, content in pegs.items():
            print(f"Колышек {name}: {content}") # Колышко : кольца на нем

        # Проверка победы
        if pegs['2'] == target_tower or pegs['3'] == target_tower:
            print("\nПОЗДРАВЛЯЮ! Ты собрала башню!")
            break

        # 3. Ход игрока
        try:
            source = input("\nОткуда снять? (1, 2, 3) или 'exit' для выхода: ").strip()
            if source.lower() == 'exit': break
            
            destination = input("Куда положить? (1, 2, 3): ").strip()

            # Проверки на правильность ввода
            if source not in pegs or destination not in pegs:
                print("Ошибка: введи номер колышка от 1 до 3!")
                continue

            if not pegs[source]:
                print("Ошибка: на этом колышке нет колец!")
                continue

            # Берем верхнее кольцо (последний элемент списка)
            top_ring = pegs[source][-1]

            # Проверяем правило "Большее на меньшее"
            # Если целевой колышек не пуст, сравниваем верхние кольца
            if pegs[destination] and top_ring > pegs[destination][-1]:
                print(f"ОШИБКА: Нельзя класть {top_ring} на {pegs[destination][-1]}")
                continue

            # 4. Перемещение
            # Снимаем с одного
            pegs[source].pop()
            # Кладем на другой
            pegs[destination].append(top_ring)

        except Exception as e:
            print(f"Произошла ошибка: {e}")
```
