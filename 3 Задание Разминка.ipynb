```python
# Даны две строки. 
# Найдите все общие уникальные символы и все символы, которые есть в первой, но нет во второй и наоборот. 
# Выведите все символы алфавита, которых нет в этих строках.
import string
set_a = {"a", "b", "c", "d", "e" }
set_b = {"c", "d", "e", "f", "g"}
alphabet = set(string.ascii_lowercase)
common = set_a.intersection(set_b)
in_a = set_a.difference(set_b)
in_b = set_b.difference(set_a)
not_in_sets = alphabet - (set_a|set_b)
print (f"Общие: {common}")
print (f"Только в первой: {in_a}")
print (f"Только во вторй: {in_b}")
print (f"Нет в строках, но есть в алфавите: {sorted(not_in_sets)}")
```

```python
# Сохраняем числа в список 
# Меняем знак, если есть отрицательное число
numbers = []

while True:
    num = int(input('Введите число (0 для выхода): '))
    if num == 0:
        break
    numbers.append (num)
print ("Исходный список", numbers)

negativ = False
for x in numbers:
    if x < 0:
        negativ = True
        break
if negativ:
    for i in range (len(numbers)):
        numbers[i] = -numbers[i]
print('Итоговый список: ', numbers)
```

```python
# Сумма элементов четной позиции / сумма элементов нечетной
list = [1, 2, 3, 4, 5, 6,]
sum_even = sum(list[0::2])
sum_odd = sum(list[1::2])
print ( sum_even/sum_odd)
```
